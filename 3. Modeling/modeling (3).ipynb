```
def prepro_csv(nums): # title_selectí•¨ìˆ˜ì—ì„œ ì•Œì•„ë‚¸ ì½”ë“œë¡œ í•´ë‹¹ ì›¹íˆ°ì˜ ëŒ“ê¸€ ì •ë³´ë§Œ ì½ì–´ë‚´ëŠ” í•¨ìˆ˜
    comment_data = pd.read_csv("webtoon_comment.csv",names =['ì›¹íˆ°ì½”ë“œ','ëŒ“ê¸€']) #csv íŒŒì¼ í˜¸ì¶œ ë° ë°ì´í„° í”„ë ˆì„í™”
    want_data = comment_data[comment_data['ì›¹íˆ°ì½”ë“œ'] == nums] # ì„ íƒí•œ ì›¹íˆ°ì˜ ëŒ“ê¸€ë§Œ ì¶”ì¶œ
    want = want_data['ëŒ“ê¸€']
    web_array = np.array(want)
    webtoon_df = []
    for i in range(len(web_array)): # ì„ íƒí•œ ëŒ“ê¸€ì„ í•˜ë‚˜ì˜ ë¦¬ìŠ¤íŠ¸ ë³€ê²½
        webtoon_df.append(web_array[i])
    return webtoon_df
```

```
def prepro_keyword(texts): # ë¬¸ì¥ ê°„ì†Œí™” í•¨ìˆ˜
    out_list = []
    for token in texts:
        remove_list = []
        remove1 = token.replace("ì„", "")
        remove2 = remove1.replace("ë¥¼","")
        remove3 = remove2.replace("[","")
        remove4 = remove3.replace("]","")
        remove5 = remove4.replace("","")
        remove6 = remove5.replace("","")
        out_list.append(remove6)
    return out_list

#ìœ„ ê³¼ì •ì€ ë°‘ì—ì„œ í‚¤ì›Œë“œë¥¼ ë½‘ì•„ë‚´ê¸° ìœ„í•´ ì‚¬ìš©í•œ summarize_with_sentenceê°€ ì˜¤ë¡œì§€ ë¬¸ì¥ë§Œ ì¸ì‹í•©ë‹ˆë‹¤. í† í¬ë‚˜ì´ì €ë¥¼ ì‚¬ìš©í•˜ë©´ ì œëŒ€ë¡œ ì¸ì‹í•˜ì§€ ëª»í•©ë‹ˆë‹¤. 
#ê¸°ì¡´ì˜ ê³¼ì •ì—ì„  ëŒ€ê´„í˜¸ì™€ í°ë”°ì˜´í‘œë¥¼ ì œê±°í•˜ëŠ”ê²Œ í˜ë“¤ì–´ì„œ í† í¬ë‚˜ì´ì €ë¥¼ ì•„ì˜ˆ í•˜ì§€ ì•Šê³  ë¬¸ì¥ ìì²´ì—ì„œ íŠ¹ì • ë‹¨ì–´ë¥¼ ì œì™¸í•˜ì˜€ìŠµë‹ˆë‹¤.
```

```
def keyword_extraction(sentences): # í‚¤ì›Œë“œ ì¶”ì¶œ í•¨ìˆ˜
    none_list = []
    try:
        penalty = lambda x:0 if (25 <= len(x) <= 80) else 1 # - penalty : íŒ¨ë„í‹° ì„¤ì • ê°€ëŠ¥
        stopwords = {'ê´€ëŒê°', 'ë„ˆë¬´', 'ì •ë§', 'ì§„ì§œ'} # - stopwords : í‚¤ì›Œë“œì—ì„œ ì œê±°ë  ë‹¨ì–´, í‚¤ì›Œë“œ ë²¡í„°ë¥¼ ë§Œë“¤ë•Œë„ ì‚¬ìš©í•˜ì§€ ì•ŠëŠ”ë‹¤.

        keywords, sents = summarize_with_sentences(
            sentences,
            penalty=penalty,
            stopwords = stopwords,
            diversity=0.5,# - diversity : ì½”ì‹¸ì¸ ìœ ì‚¬ë„ ê¸°ì¤€ í•µì‹¬ë¬¸ì¥ê°„ì˜ ìµœì†Œ ê±°ë¦¬, ê°’ì´ í´ìˆ˜ë¡ ë‹¤ì–‘í•œ ë¬¸ì¥ ì„ íƒ
            num_keywords=5, # - num_keywords : í‚¤ì›Œë“œë¡œ ì¶”ì¶œ ë  í‚¤ì›ŒìŠ¤ ê°¯ìˆ˜ë¥¼ ì„¤ì •.
            num_keysents=10, # - num_keysents : í•µì‹¬ë¬¸ì¥ìœ¼ë¡œ ì¶”ì¶œ ë  ë¬¸ì¥ì˜ ê°¯ìˆ˜ ì„¤ì •.
            verbose=False
            )
        keywords_list = []
        for i in keywords:
            keywords_list.append(i)
        return keywords_list
    except:
        return none_list
# ìœ„ì—ì„œ ë‚˜ì˜¨ í‚¤ì›Œë“œë¥¼ ì§ì ‘ ì‚¬ìš©í•˜ê¸° ìœ„í•´ì„œ ë¦¬ìŠ¤íŠ¸í™” í•˜ì˜€ìŠµë‹ˆë‹¤.
```

```
def tf(term, document): # tf-idf vectorizer í•¨ìˆ˜
    return document.count(term)

def idf(term):
    df = 0
    for doc in text:
        if term in doc:
            df = df + 1
    return log(len(text)/(df+1))+1

def tfidf(term, document):
    return tf(term, document)*idf(term)

def tfidf_list(text, keywords_list):
    tfidf_list = []
    for doc in text:
        tfidf_list.append([])
        for token in keywords_list:
            tfidf_list[-1].append(tfidf(token, doc))

    tfidf_pd = pd.DataFrame(tfidf_list, columns=keywords_list)# ê° í‚¤ì›Œë“œê°€ ë¬¸ì¥ì— ë“¤ì–´ìˆëŠ”ì§€ ë°ì´í„° ë² ì´ìŠ¤í™”
    return tfidf_list, tfidf_pd
```

```
def cos_sim(A, B): #ì½”ì‚¬ì¸ ìœ ì‚¬ë„ë¥¼ êµ¬í•˜ëŠ” í•¨ìˆ˜
       return dot(A, B)/(norm(A)*norm(B))

def cos_sim_index(tfidf_list, tfidf_pd, num): #ìš”ì²­í•œ ì›¹íˆ°ê³¼ ì½”ì‚¬ì¸ ìœ ì‚¬ë„ê°€ ê°€ì¥ ë†’ì€ ìˆœì„œë¥¼ ì°¾ì•„ë‚´ëŠ” í•¨ìˆ˜ì…ë‹ˆë‹¤.
    cos_sim_list = []
    for i in range(len(tfidf_list)):
        if i == num:
            cos_sim_list.append(0)
        else:
            cos_sim_list.append(cos_sim(tfidf_pd.values[num], tfidf_pd.values[i]))
        
    max_num = cos_sim_list.index(max(cos_sim_list))
    i_index = max_num
    print(num,'ë²ˆê³¼ ',i_index,'ë²ˆì´ ê°€ì¥ ìœ ì‚¬ ëª¨ë¸ì´ë‹¤')
    print('ì½”ì‚¬ì¸ ìœ ì‚¬ë„:', cos_sim(tfidf_pd.values[num], tfidf_pd.values[i_index]))
    return i_index
```

```
#ëŒë¦¬ëŠ”ë° 10ë¶„ ë„˜ê²Œ ê±¸ë¦¬ëŠ” ì½”ë“œì…ë‹ˆë‹¤ ì²˜ìŒ ì‹œì‘í• ë•Œ ì™¸ì—ëŠ” ëŒë¦¬ì§€ ë§ˆì„¸ìš”.
#ëŒ€ì‹  ì²˜ìŒ ì‹œì‘í• ë• ë¬´ì¡°ê±´ ëŒë ¤ì•¼ í•˜ëŠ” í•¨ìˆ˜ì…ë‹ˆë‹¤. tfidf ë°±í„°ì˜ doc_listì— í•´ë‹¹í•˜ëŠ” ë¶€ë¶„ì…ë‹ˆë‹¤.
detail_data = pd.read_csv("webtoon_detail.csv",names =['ì›¹íˆ°ì œëª©','ì ìˆ˜','ì‘ê°€','ì¥ë¥´','ì¶”ê°€ì •ë³´','ë§í¬'])
code_list = detail_data.index[1:]# ê° ì›¹íˆ°ì˜ ì½”ë“œ ë²ˆí˜¸ë§Œ ë¦¬ìŠ¤íŠ¸ì— ì…ë ¥í•©ë‹ˆë‹¤
code_nums = [] 
for code in code_list: # ì‚¬ìš©í•˜ê²Œ ì¢‹ê²Œ ë¦¬ìŠ¤íŠ¸í™” ì‹œí‚¤ëŠ” ë¶€ë¶„ì…ë‹ˆë‹¤.
    code_nums.append(int(code))

vec_list = [] 
for i in code_nums: # ë³¸ê²©ì ìœ¼ë¡œ ì›¹íˆ°ì˜ í‚¤ì›Œë“œë¥¼ í•˜ë‚˜ì˜ ë¦¬ìŠ¤íŠ¸ë¡œ í•©ì¹˜ëŠ” ê³¼ì •ì…ë‹ˆë‹¤.
    text = prepro_csv(i)
    sentence = prepro_keyword(text)
    keywords = keyword_extraction(sentence)
    forvec_list = " ".join(keywords)
    vec_list.append(forvec_list)
```
